```python
import pickle
```

```python
ent_type=dict()
with open('type.txt','r') as f:
    for line in f:
        line=line.strip().split(',')
        ent=eval(line[1])
        type=eval(line[2])
        ent_type[ent]=type
ent_type
```

```python
with open('att_data.pickle','rb') as handle:
    data=pickle.load(handle)
with open('label.pickle','rb') as handle:
    label=pickle.load(handle)
```

```python
from math import floor
pad_ent=4010
pad_rel=4011
pad_type=10
interactions=[]
for index,path_list in enumerate(data):
    inter_path_list=[]
    for path in path_list:
        inter_path=[]
        print(path)
        path_len=floor(len(path)/2)
        for i,ent in enumerate(path):
            if i%2==1:
                continue
            else:
                if i+1>=len(path):
                    inter_path.append([ent,ent_type[ent],4009])#end_rel
                else:
                    inter_path.append([ent,ent_type[ent],path[i+1]])
        while len(inter_path) <= 6:
            inter_path.append([pad_ent,pad_type,pad_rel])
        inter_path_list.append((inter_path,path_len))
    interactions.append((inter_path_list,label[index]))
    
```

```python
with open('train_interactions.txt','w') as f:
    for inter in interactions[:1800]:
        f.write(f'{inter}\n')
with open('test_interactions.txt','w') as f:
    for inter in interactions[1800:]:
        f.write(f'{inter}\n')
```

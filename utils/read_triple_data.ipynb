```python
g=list()
with open('../data/triple.txt','r') as f:
    for line in f:
        triple_list=line.strip().split(' ')
        ent1,rel,ent2=eval(triple_list[0])
        g.append((ent1,rel,ent2))
```

```python
from collections import defaultdict
ent_neighbor_dict=defaultdict(set)
```

```python
for ent1,rel,ent2 in g:
    rel=rel+4000
    ent_neighbor_dict[ent1].add((rel,ent2))
len(ent_neighbor_dict)
```

```python
ent_neighbor_dict
```

```python
import pickle
with open('ent_neighbor_dict.pickle','wb') as handle:
    pickle.dump(ent_neighbor_dict,handle)
```

```python
with open('att_data.pickle','rb') as handle:
    att_data=pickle.load(handle)
att_data
```

```python
for path_list in att_data:
    for path in path_list:
        for i,ent in enumerate(path):
            if i == len(path)-1:
                break
            if i%2==1:
                continue
            ent_neighbor_dict[path[i]].add((path[i+1],path[i+2]))
```

```python
ent_neighbor_dict
```

```python
for key,value in ent_neighbor_dict.items():
    ent_neighbor_dict[key]=list(value)
```
